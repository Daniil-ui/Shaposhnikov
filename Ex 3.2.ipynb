```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16, 12]
plt.rcParams.update({'font.size': 18})

import random


p = []

for i in range(10000):
    p.append([random.uniform(-1, 1),random.uniform(-1, 1)])
    

x = 0
for i in range(10000):
    if p[i][0]**2+p[i][1]**2 < 1:
        x = x+1
        
        
        
print('The area of our circle is ', np.pi)
print('Based of our approximation the area is', 4*x/10000)
```

```python
def f(n):
    t= []

    for i in range(10000):
     t.append(np.random.normal(-1,1,n))
            
    return t



p =[]

p.append(0)
p.append(0)
for n in range (2,10):
    p.append(f(n))


    



S=[]

#plot histograms for n values
for n in range (2,10):
    radius = []
    for i in range(10000):
            radius.append((sum(p[n][i]**2)**0.5))
    
    insidesphere = 0
    for i in range(10000):
       if radius[i] < 1:
         insidesphere = insidesphere+1
    S.append(insidesphere/10000)
    
    plt.hist(radius)
    plt.show() 
```

We notice that the histograms resemble normal distributions which makes perfect sense since we chose 10000 uniformly distributed points in R^N.

```python
x1=np.arange(2,10)
plt.plot(x1,S)
plt.axis([2, 10, 0,0.25])
plt.show()
```

We notice that the fraction of points inside the sphere looks to be exponnetialy decreasing with respect to the dimension N. This is because the colume of an N dimensional cube is 2^N while the volume of an N dimensional sphere is pi^{2/n}/Gamma(n/2+1). So the volume of the cube grows much faster than the volume of the sphere. Thus, using sampling probabilities is indeed effective in this case.
