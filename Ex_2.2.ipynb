```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [16, 8]


def a(n):
    if n > 0:
        return 8/(np.pi**2*n**2)
    else:
        return 0.25

def b(n):
    return 0


n = np.arange(0,100,1)
an = list(map(a, n))
bn = list(map(b, n))

plt.plot(n, an, color ='tab:blue') 
plt.plot(n, bn, color ='tab:orange') 
    
plt.xlim([0, 100]) 

plt.ylim([0, 1]) 

plt.xlabel("n")
plt.title('First 100 sine and cosine modes') 
    
plt.show()
```

```python
def f(x):
    if x < -1:
        return 0
    elif -1<x and x<0:
        return 1+x
    elif 0< x and x <1:
        return 1-x
    elif x > 1:
        return 0
  
    
  
def f10(x):
 res = 0
 for i in range(11):
   res = res +a(i)*np.cos((i*np.pi*x)/2)
 return res



x = np.arange(-2,2,0.01)
y = list(map(f, x))

y2 = list(map(f10,x))

plt.plot(x,y, color = 'black') 
plt.plot(x,y2, color = 'tab:orange') 
#plt.plot(x, , color ='tab:orange') 
    
plt.xlim([-2, 2]) 
plt.ylim([0, 1.8]) 


plt.title('f(x) compared to n=10 approximation') 
    
plt.show()
```
